```python
x = dff.drop(columns=(["output"]))
x
```

```python
y = dff["output"]
y
```

```python
# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()
# s_x = scaler.fit_transform(x)
# s_x
```

```python
from imblearn.over_sampling import SMOTE
s = SMOTE()
sm_x , sm_y = s.fit_resample(x , y)
```

```python
from sklearn.model_selection import train_test_split
x_tr , x_ts , y_tr , y_ts = train_test_split(x , y ,test_size = 0.2 , random_state = 1234)
```

```python
print(x_tr.shape)
print(y_tr.shape)
print(x_ts.shape)
print(y_ts.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
model = lr.fit(x_tr , y_tr)
model
```

```python
xp = model.predict(x_ts)
xp
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(xp , y_ts)
```
